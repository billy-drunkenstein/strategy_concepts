```python
from google.colab import drive
import zipfile
import os
import shutil
```

```python
drive.mount('/content/drive')
```

```python
drive_path = '/content/drive/MyDrive/Event Log/2025-7-28 申毅实习/策略研究/NLP 日度'
zip_path = os.path.join(drive_path, 'Price Data.zip')
temp_dir = '/content/temp_unzip'  # Temporary local RAM directory
final_dir = os.path.join(drive_path, 'Price Data')  # Final destination on Drive
```

```python
if os.path.exists(temp_dir):
    shutil.rmtree(temp_dir)
os.makedirs(temp_dir, exist_ok=True)
```

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(temp_dir)
```

```python
extracted_subfolder = os.path.join(temp_dir, 'Price Data')
if os.path.exists(final_dir):
    shutil.rmtree(final_dir)
shutil.move(extracted_subfolder, final_dir)
```
