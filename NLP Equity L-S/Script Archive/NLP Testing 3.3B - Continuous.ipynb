```python
device = 0 if torch.cuda.is_available() else -1
print("Device set to", "GPU" if device == 0 else "CPU")
```

```python
nllb_model_name = "facebook/nllb-200-3.3B"
nllb_tokenizer = AutoTokenizer.from_pretrained(nllb_model_name)
nllb_model = AutoModelForSeq2SeqLM.from_pretrained(nllb_model_name)

translator = pipeline(
    "translation",
    model=nllb_model,
    tokenizer = nllb_tokenizer,
    src_lang = "zho_Hans",
    tgt_lang = "eng_Latn",
    device=device,
    truncation = True,
    max_length = 512
)
```

```python
finbert_model = "ProsusAI/finbert"
tokenizer = AutoTokenizer.from_pretrained(finbert_model)
model = AutoModelForSequenceClassification.from_pretrained(finbert_model)

finbert_regressor = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    device=device
)
```

```python
equity_reports = pd.read_csv(
    'equity_reports.csv',
    encoding='utf-8',
    parse_dates=['PUBLISH_DATE', 'UPDATE_TIME', 'UPDATE_DATE']
)
```

```python
sentiment_map = {
    'POSITIVE': '正面',
    'NEUTRAL': '中性',
    'NEGATIVE': '负面'
}

verbose = True
```

```python
start_time = time.time()

for i, row in equity_reports.head(10).iterrows():
    if verbose:
        print(f"\n=== Row {i + 1} ===")

    title_cn = row['TITLE']
    summary_cn = row['SUMMARY']

    try:
        if pd.notna(title_cn):
            title_en = translator(title_cn, max_length=512, truncation=True)[0]['translation_text']
            title_score = finbert_regressor(title_en)[0]['score']
            if verbose:
                print(f"TITLE: {title_cn[:150]}{'...' if len(title_cn) > 150 else ''}")
                print(f"  ↳ {title_en[:100]}{'...' if len(title_en) > 100 else ''}")
                print(f"  → 情绪得分: {title_score:.3f}")
        else:
            if verbose:
                print("TITLE: [缺失]")

        if pd.notna(summary_cn):
            summary_en = translator(summary_cn, max_length=512, truncation=True)[0]['translation_text']
            summary_score = finbert_regressor(summary_en)[0]['score']
            if verbose:
                print(f"SUMMARY: {summary_cn[:150]}{'...' if len(summary_cn) > 150 else ''}")
                print(f"  ↳ {summary_en[:100]}{'...' if len(summary_en) > 100 else ''}")
                print(f"  → 情绪得分: {summary_score:.3f}")
        else:
            if verbose:
                print("SUMMARY: [缺失]")

    except Exception as e:
        print(f"Error on row {i}: {e}")

end_time = time.time()
```

```python
elapsed = end_time - start_time
print(f"\nProcessed 100 rows in {elapsed:.2f} seconds")
print(f"Average time per row: {elapsed / 100:.2f} seconds")
```
