```python
from PIL import Image
import os
```

```python
with open("指数清单/最新分表/latest_date.txt", "r", encoding="utf-8") as f:
    date_str = f.read().strip()  # yyyy-mm-dd

date_str
```

```python
input_dir = os.path.join("指数清单", "最新分表")

output_dir = os.path.join("指数清单", date_str)
os.makedirs(output_dir, exist_ok = True)
```

```python
image_dir = "指数清单/最新分表"

groups = {
    "A股多投": ["A股多投-1年.png", "A股多投-3年.png"],
    "A股空投": ["A股空投-1年.png", "A股空投-3年.png"],
    "海外多投": ["海外多投-1年.png", "海外多投-3年.png"],
    "海外空投": ["海外空投-1年.png", "海外空投-3年.png"],

    # ==== 新增: 当前值 ====
    "A股 当前": ["A股 当前 多投.png", "A股 当前 空投.png"],
    "海外 当前": ["海外 当前 多投.png", "海外 当前 空投.png"],
}

gap = 200
```

```python
for name, files in groups.items():
    imgs = [Image.open(os.path.join(input_dir, f)) for f in files]

    widths = [img.width for img in imgs]
    heights = [img.height for img in imgs]

    total_height = sum(heights) + gap * (len(imgs) - 1)
    max_width = max(widths)

    new_img = Image.new("RGB", (max_width, total_height), (255, 255, 255))

    y = 0
    for i, img in enumerate(imgs):
        new_img.paste(img, (0, y))
        y += img.height + gap

    new_img.save(os.path.join(output_dir, f"{name}.png"))
```
